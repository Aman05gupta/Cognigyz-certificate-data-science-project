### Explore the distribution of categorical variables like "Country Code," "City," and "Cuisines."

```python
plt.figure(figsize=(10,7))
sns.countplot(x="Country Code", data=df)

plt.title("Distribution of Restaurants by Country Code")
plt.xlabel("Country Code")
plt.ylabel("Number of Restaurants")
plt.show()
```

```python
top_Countries = df["Country Code"].value_counts().head()
print("Sort and display the top 5 countries with the most restaurants : ")
print(top_Countries)
```

```python
# Get the top 20 cities by restaurant count
top_cities = df['City'].value_counts().head(20).index

# Filter dataframe for the top cities
filtered_df = df[df['City'].isin(top_cities)]

# Plot the distribution
plt.figure(figsize=(17,8))
sns.countplot(x='City', data=filtered_df, hue='City', palette="Set2", legend=False)
plt.title("Distribution of Restaurants by City")
plt.xlabel("City")
plt.ylabel("Number of Restaurants")
plt.xticks(rotation=45)
plt.show()
```

```python

plt.figure(figsize=(17,8))

# Get the top 20 cuisines by count
Cuisines_count = df['Cuisines'].value_counts()


Cuisines_count.head(20).plot(kind="bar", color=sns.color_palette("Set2"))


plt.title("Top 20 Cuisines with the Highest Number of Restaurants")
plt.xlabel("Cuisines")
plt.ylabel("Number of Restaurants")
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python
top_cities = df["City"].value_counts().head(10)
print("Top 10 cites with the Highest Number of Restaurants : ")
print(top_cities)
```

```python
top_cuisines = Cuisines_count.head(10)
print("Top 10 Cuisines with the Highest Number of Restaurants :")
print(top_cuisines)
```
